```python
import matplotlib.pyplot as plt
import pandas as pd


try:
    file_path = '/content/한국도로공사_교통사고통계_20241231.csv'
    df = pd.read_csv(file_path, encoding='cp949')
except UnicodeDecodeError:
    df = pd.read_csv(file_path, encoding='euc-kr')
except FileNotFoundError:
    print("파일을 찾을 수 없습니다. 경로를 확인해 주세요.")


years = df['연도']
accidents = df['사고']


years_recent = years.tail(4)
accidents_recent = accidents.tail(4)
car_registrations_recent = [2491, 2546, 2594, 2630]

fig, ax1 = plt.subplots(figsize=(8, 5))


line1 = ax1.plot(years_recent, accidents_recent, marker='o', color='b', label='Number of traffic accidents')
ax1.set_title('Trends in the number of traffic accidents and registered vehicles by year (last 4 years)', pad=15)
ax1.set_xlabel('year')
ax1.set_ylabel(' number of incidents', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.set_xticks(years_recent)


ax2 = ax1.twinx()
line2 = ax2.plot(years_recent, car_registrations_recent, marker='s', color='orange', linestyle='--', label='Number of registered cars')
ax2.set_ylabel('Number of registered vehicles (in ten thousand)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')


lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left')

ax1.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
